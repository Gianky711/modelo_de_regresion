ANALISIS DE ERRORES

```python
x_entrenamiento, x_prueba, y_entrenamiento, y_prueba = train_test_split(X, y)
```

```python
# Crear y ajustar el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(x_entrenamiento, y_entrenamiento)
```

```python
# Realizar predicciones en el conjunto de prueba
y_pred = modelo.predict(x_prueba)
# Visualizar las predicciones
print("Predicciones:", y_pred)
```

```python
# Calcular el MSE y R^2
mse = mean_squared_error(y_prueba, y_pred)
r2 = r2_score(y_prueba, y_pred)
```

```python
print(f'Mean Squared Error (MSE): {mse}')
print(f'Coeficiente de Determinación (R^2): {r2}')
```

ENTRENAMIENTO DEL PROGRAMA

```python
# Seleccionar variables independientes (X) y variable dependiente (y)
X = D[['Resultados', 'Tiempo']]
y = D['Costo_por_resultados']
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
x_entrenamiento, x_prueba, y_entrenamiento, y_prueba = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Imprimir las formas de los conjuntos de entrenamiento y prueba
print("Forma de X_train:", x_entrenamiento.shape)
print("Forma de X_test:", x_prueba.shape)
print("Forma de y_train:", y_entrenamiento.shape)
print("Forma de y_test:", y_prueba.shape)
```

```python
# Rendimiento o performance del modelo 100%
modelo.score(x_prueba, y_prueba) *100
```
